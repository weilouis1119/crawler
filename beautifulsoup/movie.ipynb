```python
import requests,bs4,os
```

```python
url ='https://movies.yahoo.com.tw/movie_thisweek.html'
res = requests.get(url)
bs = bs4.BeautifulSoup(res.text, 'lxml')
infos = bs.find('ul', class_ = 'release_list').find_all('li')

photo = []
for info in infos:
    chinese_name = info.find('div',class_='release_info').find('a').text
    eng_name = info.find('div',class_='en').find('a').text
    time = info.find('div',class_ = 'release_movie_time').text
    expect = info.find('div',class_ = 'leveltext').find('span').text
    article = info.find('div',class_ = 'release_text').find('span').text
    img = info.find('img')['src']

    photo.append(img)
    pre = info.find('div', class_ = 'release_btn color_btnbox').find_all('a')[1].get('href')
    print('中文名稱:',chinese_name.replace(' ','').replace('\n',''))
    print('英文名稱:',eng_name.replace(' ','').replace('\n',''))
    print(time.replace(' ','').replace('\n',''))
    print('期待度:',expect)
    print('摘要:',article.replace(' ',''))
    print('圖片:',img)
    print('預告片:',pre)
    print('==========================================')
```

```python
#儲存圖片
destDir = 'images_movie'
if os.path.exists(destDir) == False:
    os.mkdir(destDir)
for i in photo:
    picture = requests.get(i)
    picture.raise_for_status()
    print('海報下載成功')
    pictFile = open(os.path.join(destDir,os.path.basename(i)),'wb')
    for diskStorage in picture.iter_content(10240):
        pictFile.write(diskStorage)
    pictFile.close()
```
